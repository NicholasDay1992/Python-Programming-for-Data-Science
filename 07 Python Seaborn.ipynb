## `sns.scatter()`

* We can't use exactly the same arguments as we did for plt...

```python
x = np.arange(1, 10, 1)
y = x**2 
sns.scatterplot(x, y)
```

We could resolve this by bundling them as as a `tuple`: `(x, y)`

```python
x = np.arange(1, 10, 1)
y = x**2 
sns.scatterplot(data = (x, y))
```

But notice that this draws two sets of observations... x vs x, and y vs x !

If we want one dataset: x vs y, we can use the `x` and `y` arguments: 

```python
x = np.arange(1, 10, 1)
y = x**2 
sns.scatterplot(x=x, y=y)
```

## `sns.regplot()` 

* draws a regression line of best fit - same as `np.polyfit()`

```python
sns.regplot(x = x, y = x)
```

```python
sns.regplot(x = x, y = y)
```

## `np.polyfit()` 

* line of best fit - for drawing on plt plots

```python
a, b = np.polyfit(x, y, 1)
plt.plot(x, a*x+b) 
```

## What about quadratic and polynomial data?

```python
x1 = x
x2 = x
y1 = x**2
y2 = x**3
```

## `plt.plot()` can draw us a polynomial regression curve

```python
plt.plot(x,y)
```

```python
sns.scatterplot(x=x1, y=y1)
sns.scatterplot(x=x2, y=y2)
```

```python
sns.regplot(x=x1, y=y1, scatter=False)
sns.regplot(x=x2, y=y2, scatter=False)
```

```python
sns.regplot(x=x1, y=y1, scatter=True, label="X Squared")
sns.regplot(x=x2, y=y2, scatter=True, label="X Cubed")

plt.legend()
```

## `sns.regplot()` doesn't naturally handle polynomial regression

```python
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [1.2, 1.9, 3.2, 3.8, 5.1]
})

df['x2'] = df['x'] ** 2

# Use regplot with x as the polynomial-transformed variable
sns.regplot(x='x', y='y', data=df, order=2)  # This works only for regplot via keyword `order`

plt.show()
```
